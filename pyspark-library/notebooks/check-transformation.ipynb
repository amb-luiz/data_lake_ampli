# Este notebook serve para desenvolvimento local de transformações em sparksql

## 1- Criando Spark magic com o alias %sql

```python
spark
```

```python
%load_ext sparksql_magic
```

```python
%alias_magic sql sparksql
```

## 2- Read local input

```python
df_json = spark.read.format('json').option('Multiline','true').load('/spark-pyutil/data/input/json')
```

```python
df_json.take(10)
```

```python
df_json.count()
```

```python
df_json.createOrReplaceTempView('json')
```

```python
%%sql
select field1 from json where limit 10
```

## 2- Read local output

```python
df_json_output = spark.read.format('parquet').load('/spark-pyutil/data/output/json')
```

```python
df_json_output.createOrReplaceTempView('json_output')
```

```python
%%sql
select *  from json_output limit 10
```
